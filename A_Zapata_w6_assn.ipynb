```python
#first import pandas
import pandas as pd

#load raw file from github
filename = "https://raw.githubusercontent.com/AlejandraZS10/diamonds_data/master/diamonds.csv"
diamonds = pd.read_csv(filename)

#use the head function to make sure the data loaded properly
diamonds.head()
```

### 1. What proportion of diamonds are between .30 and 1.08 carats?

I looked at the variables to create the proportion calculation.

```python
carats_total=len(diamonds.carat) #the total of rows in carats

carat_range=diamonds.carat[(diamonds.carat >= 0.30) & (diamonds.carat <= 1.08)].count() 
#this variable counts only those diamonds with a carat within the range

proportion = round((carat_range/carats_total), 2)
#calculates the proportion

print('The diamond count between 0.30 and 1.08 is ' + str(carat_range))
print('The diamond count is ' + str(carats_total))

print('The proportion of diamonds between .30 and 1.08 is ' + str(proportion))
       
```

### 3. How many of the diamonds have a carat less than the mean carat value?

Caculate the mean using .mean() and then compare the given float with the column.
Count how many elements are true to this condition.

```python
carat_mean = round((diamonds.carat.mean()),2) #calculates the mean carat

less_carat_mean = diamonds.carat[diamonds.carat < carat_mean].count() #variable that counts only diamonds with a lower carat value

print('The mean carat is ' +str(carat_mean))

print('There are ' + str(less_carat_mean) + ' with a lower carat than the mean')
```

### 5. Which diamond has the highest price per carat? What is its value?

Create a new column we can later refer to of price per carat. 
Obtain the max value using max().

```python
#Adds a new column with the price per carat
diamonds["pricepercarat"] = (diamonds["price"] / diamonds["carat"])

diamonds.head()
```

```python
max_price=(round(max(diamonds.pricepercarat),2)) #gives the max mean value of the new column

print('The diamond with the highest price/carat is ' +str(max_price))
```
